```python
## Import Python libraries
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
ApplianceShipments_df = pd.read_csv('ApplianceShipments.csv')
```

```python
ApplianceShipments_df = ApplianceShipments_df.iloc[0:,1:2]
date_range = pd.date_range(start='1/1/1985', end='12/31/1989', freq='Q')
ApplianceShipments_df['Date'] = pd.to_datetime(date_range)
```

```python
Shipments_ts = pd.Series(ApplianceShipments_df.Shipments.values,index=ApplianceShipments_df.Date)
```

```python
Shipments_ts.plot(ylim=[3000, 5500], legend=False)
plt.xlabel('Year') # set x-axis label
plt.ylabel('Shipments') # set y-axis label
```
